```python
import numpy as np
from matplotlib import pyplot as plt

np.random.seed(14)
```

```python
std = 1
array_dim = 10000

normal_array1 = np.random.randn(1, array_dim)
normal_array2 = np.random.randn(1, array_dim)

rayleigh_array = std * np.sqrt(normal_array1**2 + normal_array2**2)

rayleigh_mean = np.mean(rayleigh_array)
empirical_rayleigh_mean = std * np.sqrt(np.pi / 2) 

rayleigh_var = np.var(rayleigh_array)
empirical_rayleigh_var = std**2 * (4 - np.pi) / 2
```

```python
rayleigh_prob = (rayleigh_array / std**2) * np.exp((-rayleigh_array**2) / (2*std**2))
```

```python
counts, bin_edges = np.histogram(rayleigh_array, bins=60)

probabilities = counts / counts.sum()

bin_centers = (bin_edges[:-1] + bin_edges[1:]) / 2

plt.bar(bin_centers, probabilities, width=bin_edges[1] - bin_edges[0], edgecolor='k', alpha=0.7)

#plt.plot(rayleigh_array, rayleigh_prob, 'x')

plt.show()
```
