## Uniform Distribution

### pdf

$$ p_{X} (x) = \frac{1}{b - a}, \quad a \le x \le b $$

### Mean

$$ \mathbb{E} \{X \} = \frac{a + b}{2} $$

### Variance

$$ \sigma_{X}^2 = \frac{(b - a)^2}{12} $$ 

```python
import numpy as np
from matplotlib import pyplot as plt

# Setting the interval [a, b] as [1, 10]

a = 1
b = 10
pdf = 1 / (b - a) # As seen in equation above, pdf is constant for any value of x in interval [a, b]


# Setting the axis

x_values = np.array([point for point in range(a, b+1)]) # Array of 10 values in [a, b]

pdf_values = np.array([pdf for point in range(a, b+1)]) # Array of pdf values for each point of x  


# Plotting the curve

plt.plot(x_values, pdf_values, color='black')

plt.xlim(0, 11)
plt.ylim(0, 0.15)


plt.vlines(a, ymin=0, ymax=pdf, colors='red', linestyle='dashed')
plt.vlines(b, ymin=0, ymax=pdf, colors='red', linestyle='dashed')

plt.title('Uniform Distribution')
plt.xlabel('x')
plt.ylabel('pdf')

plt.grid()
```

### Exponential Distribution

### pdf

$$ p_{X} (x) = \lambda \exp(-\lambda x), \quad a \ge 0, \quad \lambda \ge 0 $$

### Mean

$$ \mathbb{E} \{X \} = \frac{1}{\lambda} $$

### Variance

$$ \sigma_{X}^2 = \frac{1}{\lambda^2} $$ 

```python
# Setting the interval [a, b] as [0, 3] (although it should be [0, inf))

a = 0
b = 3
lamb = 3


# Setting pdf 

def pdf_exp(lamb_, x_value):

    return lamb_ * np.exp(- lamb_ * x_value) 


# Setting the axis

x_values = np.linspace(a, b, num=100) # Array of 100 values in [a, b]

pdf_values = pdf_exp(lamb, x_values) # Array of pdf values for each point of x  


# Plotting the curve

plt.plot(x_values, pdf_values, color='black')

plt.xlim(0, 3)
plt.ylim(0, 3)


# Setting the mean

mean = 1 / lamb

plt.vlines(mean, ymin=0, ymax=pdf_exp(lamb, mean), colors='red', linestyle='dashed')


plt.title('Exponential Distribution')
plt.xlabel('x')
plt.ylabel('pdf')

plt.grid()
```

### Gaussian Distribution

### pdf

$$ p_{X} (x) = \frac{1}{\sqrt{2 \pi} \sigma} \exp(-\frac{(x - \mu)^2}{2 \sigma^2}), \quad - \infty \lt x \lt \infty, \quad \sigma \gt 0 $$

### Mean

$$ \mathbb{E} \{X \} = \mu $$

### Variance

$$ \sigma_{X}^2 = \sigma^2 $$ 

```python
# Setting the interval [a, b] as [-8, 9] (although it should be (inf, inf))

a = -8
b = 9
sigma = 2
mu = 1

# Setting pdf 

def pdf_gauss(mu_, sigma_, x_value):

    return (1 / (np.sqrt(2 * np.pi) * sigma_)) * np.exp(- (x_value - mu_)**2 / (2 * sigma_**2)) 


# Setting the axis

x_values = np.linspace(a, b, num=100) # Array of 100 values in [a, b]

pdf_values = pdf_gauss(mu, sigma, x_values) # Array of pdf values for each point of x  


# Plotting the curve

plt.plot(x_values, pdf_values, color='black')

plt.xlim(-8, 10)
plt.ylim(0, 0.2)


# Setting the mean

mean = mu

plt.vlines(mean, ymin=0, ymax=pdf_gauss(mu, sigma, mean), colors='red', linestyle='dashed')


plt.title('Gaussian Distribution')
plt.xlabel('x')
plt.ylabel('pdf')

plt.grid()
```

### Gamma Distribution

### pdf

$$ p_{X} (x) = \frac{1}{\lambda^{\alpha} \Gamma(\alpha)} x^{\alpha-1} \exp(-\frac{x}{\lambda}), 
\quad x \gt 0, \quad \alpha \gt 0, \quad \lambda \gt 0 $$

### Mean

$$ \mathbb{E} \{X \} = \alpha \cdot \lambda $$

### Variance

$$ \sigma_{X}^2 = \alpha \cdot \lambda^2 $$ 

```python
# Setting the interval [a, b] as [0, 50] (although it should be (0, inf))

a = 0
b = 50
alpha = 3
lamb = 5


# Setting gamma function as gamma(n) = (n - 1)!

def gamma(n):

    product = 1
    
    for num in range(1, n):

        product *= num

    return product


# Setting pdf 

def pdf_gamma(alpha_, lamb_, x_value):

    return (1 / (lamb_**alpha_ * gamma(alpha_))) * x_value**(alpha_ - 1) * np.exp(- x_value / lamb_) 


# Setting the axis

x_values = np.linspace(a, b, num=100) # Array of 100 values in [a, b]

pdf_values = pdf_gamma(alpha, lamb, x_values) # Array of pdf values for each point of x  


# Plotting the curve

plt.plot(x_values, pdf_values, color='black')

plt.xlim(0, 50)
plt.ylim(0, 0.06)


# Setting the mean

mean = alpha * lamb

plt.vlines(mean, ymin=0, ymax=pdf_gamma(alpha, lamb, mean), colors='red', linestyle='dashed')


plt.title('Gamma Distribution')
plt.xlabel('x')
plt.ylabel('pdf')

plt.grid()
```

### Rayleigh Distribution

### pdf

$$ p_{X} (x) = \frac{x}{\sigma^2} \exp(-\frac{x^2}{2 \sigma^2}), 
\quad x \ge 0, \quad \sigma \gt 0 $$

### Mean

$$ \mathbb{E} \{X \} = \sigma \sqrt{\frac{\pi}{2}}$$

### Variance

$$ \sigma_{X}^2 = (\frac{4 - \pi}{2}) \sigma^2 $$ 

```python
# Setting the interval [a, b] as [0, 3] (although it should be [0, inf))

a = 0
b = 3
sigma = 0.5


# Setting pdf 

def pdf_rayleigh(sigma_, x_value):

    return (x_value / sigma_**2) * np.exp(- (x_value**2) / (2*sigma**2)) 


# Setting the axis

x_values = np.linspace(a, b, num=100) # Array of 100 values in [a, b]

pdf_values = pdf_rayleigh(sigma, x_values) # Array of pdf values for each point of x  


# Plotting the curve

plt.plot(x_values, pdf_values, color='black')

plt.xlim(0, 3)
plt.ylim(0, 1.4)


# Setting the mean

mean = sigma * np.sqrt(np.pi / 2)

plt.vlines(mean, ymin=0, ymax=pdf_rayleigh(sigma, mean), colors='red', linestyle='dashed')


plt.title('Rayleigh Distribution')
plt.xlabel('x')
plt.ylabel('pdf')

plt.grid()
```

### Cauchy Distribution

### pdf

$$ p_{X} (x) = \frac{\frac{\alpha}{\pi}}{x^2 + \alpha^2} , 
\quad - \infty \le x \le  \infty, \quad \alpha \gt 0 $$

### Mean and variance are not defined

```python
# Setting the interval [a, b] as [-5, 5] (although it should be (-inf, inf))

a = -5
b = 5
alpha = 0.5


# Setting pdf 

def pdf_cauchy(alpha_, x_value):

    return (alpha_ / np.pi) / (x_value**2 + alpha_**2) 


# Setting the axis

x_values = np.linspace(a, b, num=1000) # Array of 1000 values in [a, b]

pdf_values = pdf_cauchy(alpha, x_values) # Array of pdf values for each point of x  


# Plotting the curve

plt.plot(x_values, pdf_values, color='black')

plt.xlim(-5, 5)
plt.ylim(0, 0.7)


plt.title('Cauchy Distribution')
plt.xlabel('x')
plt.ylabel('pdf')

plt.grid()
```

### Laplace Distribution

### pdf

$$ p_{X} (x) = \frac{\alpha \exp(-\alpha |x|)}{2}, 
\quad - \infty \lt x \lt \infty, \quad \alpha \gt 0 $$

### Mean

$$ \mathbb{E} \{X \} = 0 $$

### Variance

$$ \sigma_{X}^2 = \frac{2}{\alpha^2} $$ 

```python
# Setting the interval [a, b] as [-5, 5] (although it should be (-inf, inf))

a = -5
b = 5
alpha = 2


# Setting pdf 

def pdf_laplace(alpha_, x_value):

    return (alpha_ * np.exp(-alpha_ * np.absolute(x_value))) / 2


# Setting the axis

x_values = np.linspace(a, b, num=10000) # Array of 100 values in [a, b]

pdf_values = pdf_laplace(alpha, x_values) # Array of pdf values for each point of x  


# Plotting the curve

plt.plot(x_values, pdf_values, color='black')

plt.xlim(-5, 5)
plt.ylim(0, 1)


# Setting the mean

mean = 0

plt.vlines(mean, ymin=0, ymax=pdf_laplace(alpha, mean), colors='red', linestyle='dashed')


plt.title('Laplace Distribution')
plt.xlabel('x')
plt.ylabel('pdf')

plt.grid()
```

### Poisson Distribution

### pdf

$$ Pr(k) = \frac{\alpha^k}{k!} \exp(- \alpha), 
\quad k = 0, 1, 2, \dots, \quad \alpha \gt 0 $$

### Mean

$$ \mathbb{E} \{X \} = \alpha $$

### Variance

$$ \sigma_{X}^2 = \alpha $$ 

```python
# Setting the interval [a, b] as [1, 40]

a = 1
b = 40
alpha_1 = 10
alpha_2 = 20


# Setting factorial function as factorial(n) = n!

def factorial(n):

    product = 1
    
    for num in range(1, n+1):

        product *= num

    return product


# Setting pdf 

def pdf_poisson(alpha_, x_value):

    poisson = np.zeros(x_value.shape[0])

    for value in range(x_value.shape[0]):

        poisson[value] = ((alpha_**x_value[value]) / (factorial(int(x_value[value])))) * np.exp(-alpha_) 

    return poisson


# Setting the axis

x_values = np.linspace(a, b, num=b-a+1) # Array of values of k

pdf_values_1 = pdf_poisson(alpha_1, x_values) # Array of pdf values for each point of x for alpha = 10
pdf_values_2 = pdf_poisson(alpha_2, x_values) # Array of pdf values for each point of x for alpha = 20


# Plotting the curve

plt.plot(x_values, pdf_values_1, 'ko', label='alpha = 10')
plt.plot(x_values, pdf_values_2, 'rs', label='alpha = 20')

for oc in range(x_values.shape[0]):

    plt.plot((x_values[oc], x_values[oc]), (0, pdf_values_1[oc]), 'k-')
    plt.plot((x_values[oc], x_values[oc]), (0, pdf_values_2[oc]), 'r-')


plt.xlim(0, 40)
plt.ylim(0, 0.14)

plt.title('Poisson Distribution')
plt.xlabel('k')
plt.ylabel('pdf')
plt.legend()

plt.grid()
```
